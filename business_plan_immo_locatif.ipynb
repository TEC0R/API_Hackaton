```python
import math

print("ğŸ¦ PROFIL DE Lâ€™INVESTISSEUR ET STRUCTURE DE LA SCI")

avec_credit = input("ğŸ‘‰ As-tu un crÃ©dit immobilier pour ce bien ? (oui/non) : ").strip().lower()
apport = 0.0
taux_annuel = 0.0
duree_annees = 0
assurance_credit = 0.0
mensualite_credit = 0.0
mensualite_assurance_credit = 0.0
interets_emprunt = 0.0
capital_emprunte = 0.0

if avec_credit == "oui":
    apport = float(input("ğŸ‘‰ Montant de ton apport personnel (â‚¬) : "))
    taux_annuel = float(input("ğŸ‘‰ Taux dâ€™intÃ©rÃªt annuel du prÃªt (%) : ")) / 100
    duree_annees = int(input("ğŸ‘‰ DurÃ©e totale du prÃªt (en annÃ©es) : "))
    assurance_credit = float(input("ğŸ‘‰ Taux dâ€™assurance du prÃªt (%/an du capital empruntÃ©) : ")) / 100

print("\nğŸ¢ GESTION ET ASSURANCE DU BIEN")
gestion_agence = input("ğŸ‘‰ Gestion locative confiÃ©e Ã  une agence ? (oui/non) : ").strip().lower()
frais_agence_pct = 0.0
pno_incluse = "non"
assurance_bien = 0.0

if gestion_agence == "oui":
    frais_agence_pct = float(input("ğŸ‘‰ Taux de frais de gestion de lâ€™agence (%) : ")) / 100
    pno_incluse = input("ğŸ‘‰ Lâ€™agence inclut-elle une assurance PNO ? (oui/non) : ").strip().lower()
    if pno_incluse == "non":
        assurance_bien = float(input("ğŸ‘‰ Montant annuel de la PNO (â‚¬) : "))
else:
    assurance_bien = float(input("ğŸ‘‰ Montant annuel de la PNO (â‚¬) : "))

autres_depenses = float(input("ğŸ‘‰ Autres dÃ©penses diverses (entretien, comptable, etc.) (â‚¬) : "))

print("\nğŸ›ï¸ RÃ‰GIME FISCAL DE LA SCI")
regime_fiscal = input("ğŸ‘‰ RÃ©gime fiscal ('ir' ou 'is') : ").strip().lower()

# ParamÃ¨tres selon le rÃ©gime
taux_is = 0.0
part_terrain = 0.0
duree_amortissement = 0
revenu_net = 0.0
parts = 1.0

if regime_fiscal == "is":
    taux_is = float(input("ğŸ‘‰ Taux dâ€™IS (%) (ex : 15 ou 25) : ")) / 100
    part_terrain = float(input("ğŸ‘‰ Pourcentage du prix correspondant au terrain (%) : ")) / 100
    duree_amortissement = int(input("ğŸ‘‰ DurÃ©e dâ€™amortissement du bÃ¢ti (en annÃ©es) : "))
else:
    revenu_net = float(input("ğŸ‘‰ Revenu NET imposable hors SCI (â‚¬) : "))
    parts = float(input("ğŸ‘‰ Nombre de parts fiscales du foyer : "))
```

```python
print("\nğŸ  INFORMATIONS DU BIEN Ã€ SIMULER")

prix_bien = float(input("ğŸ‘‰ Prix dâ€™achat du bien (â‚¬) : "))
if avec_credit == "oui":
    capital_emprunte = prix_bien - apport
    taux_mensuel = taux_annuel / 12
    nb_mois = duree_annees * 12
    mensualite_credit = capital_emprunte * (taux_mensuel / (1 - (1 + taux_mensuel) ** -nb_mois))
    mensualite_assurance_credit = (capital_emprunte * assurance_credit) / 12
    interets_emprunt = capital_emprunte * taux_annuel / 2  # approximation moyenne
else:
    apport = prix_bien

loyer_mensuel = float(input("ğŸ‘‰ Loyer mensuel perÃ§u (â‚¬) : "))
charges_bien = float(input("ğŸ‘‰ Charges annuelles non rÃ©cupÃ©rables (â‚¬) : "))

print("\nğŸ™ï¸ TAXE FONCIÃˆRE")
mode_tf = input("ğŸ‘‰ Souhaites-tu entrer un montant exact (1) ou faire une estimation (2) ? : ").strip()
if mode_tf == "1":
    taxe_fonciere = float(input("ğŸ‘‰ Montant annuel exact de la taxe fonciÃ¨re (â‚¬) : "))
else:
    surface = float(input("ğŸ‘‰ Surface habitable du bien (mÂ²) : "))
    type_bien = input("ğŸ‘‰ Type de bien ('appartement' ou 'maison') : ").strip().lower()
    taux_communal = float(input("ğŸ‘‰ Taux global de taxe fonciÃ¨re (%) : ")) / 100
    tarif_m2 = 150 if type_bien == "appartement" else 130
    taxe_fonciere = surface * tarif_m2 * 0.5 * taux_communal

travaux_deductibles = float(input("ğŸ‘‰ Montant des travaux de rÃ©paration/amÃ©lioration (â‚¬) : "))
travaux_energetiques = input("ğŸ‘‰ Travaux Ã©nergÃ©tiques (oui/non) : ").strip().lower()

# ================================================================
# ğŸ“Š CALCULS AUTOMATIQUES ET FISCAUX
# ================================================================
loyers_annuels = loyer_mensuel * 12
frais_agence_annuels = loyers_annuels * frais_agence_pct
depenses_annexes = taxe_fonciere + assurance_bien + autres_depenses + frais_agence_annuels

if regime_fiscal == "is":
    base_amortissable = prix_bien * (1 - part_terrain)
    amortissement_annuel = base_amortissable / duree_amortissement
    benefice_imposable = loyers_annuels - (
        charges_bien + depenses_annexes + interets_emprunt + amortissement_annuel
    )
    cout_fiscal_total = max(0, benefice_imposable * taux_is)
    revenu_locatif_imposable = benefice_imposable
    prelevements_sociaux = 0
    surcout_ir = 0
else:
    plafond_deficit = 21400 if travaux_energetiques == "oui" else 10700
    autres_charges_deductibles = charges_bien + depenses_annexes + travaux_deductibles
    revenu_locatif_imposable = loyers_annuels - (interets_emprunt + autres_charges_deductibles)

    if revenu_locatif_imposable < 0:
        deficit_total = abs(revenu_locatif_imposable)
        imputable_global = min(deficit_total, plafond_deficit)
        revenu_locatif_imposable = 0
        deficit_foncier = imputable_global
    else:
        deficit_foncier = 0

    revenu_total = revenu_net + max(0, revenu_locatif_imposable) - deficit_foncier

    def impot_progressif(revenu, parts):
        tranches = [
            (0, 11294, 0.0),
            (11294, 28797, 0.11),
            (28797, 82341, 0.30),
            (82341, 177106, 0.41),
            (177106, math.inf, 0.45),
        ]
        rp = revenu / parts
        imp = 0
        for bas, haut, taux in tranches:
            if rp > bas:
                imp += (min(rp, haut) - bas) * taux
            else:
                break
        return imp * parts

    impot_avant = impot_progressif(revenu_net, parts)
    impot_apres = impot_progressif(revenu_total, parts)
    surcout_ir = max(0, impot_apres - impot_avant)
    prelevements_sociaux = max(0, revenu_locatif_imposable * 0.172)
    cout_fiscal_total = surcout_ir + prelevements_sociaux

# ================================================================
# ğŸ’° SYNTHÃˆSE
# ================================================================
cout_mensuel_total = (
    mensualite_credit
    + mensualite_assurance_credit
    + (charges_bien / 12)
    + (depenses_annexes / 12)
    + (cout_fiscal_total / 12)
)

print("\n=================== ğŸ’° RÃ‰SULTATS ===================")
print(f"RÃ©gime fiscal : SCI Ã  {regime_fiscal.upper()}")
print(f"Revenu locatif imposable : {revenu_locatif_imposable:,.2f} â‚¬/an")
print(f"SurcoÃ»t IR : {surcout_ir:,.2f} â‚¬/an")
print(f"PrÃ©lÃ¨vements sociaux : {prelevements_sociaux:,.2f} â‚¬/an")
print(f"CoÃ»t fiscal total : {cout_fiscal_total:,.2f} â‚¬/an")
print(f"â¡ï¸ ğŸ’¸ CoÃ»t total global : {cout_mensuel_total:,.2f} â‚¬/mois")

# ---------------------------
# Ã‰TAPE 8 â€” PRIX Dâ€™ACHAT MAXIMUM Ã€ Lâ€™Ã‰QUILIBRE
# ---------------------------
print("\nğŸ“ˆ CALCUL DU PRIX Dâ€™ACHAT MAXIMUM Ã€ Lâ€™Ã‰QUILIBRE")

if avec_credit == "oui" and mensualite_credit > 0:
    loyer_mensuel_net = loyer_mensuel

    # --- CoÃ»ts mensuels selon diffÃ©rents scÃ©narios ---
    cout_fiscal_mensuel = cout_fiscal_total / 12.0
    charges_mensuelles = charges_bien / 12.0
    depenses_annexes_mensuelles = depenses_annexes / 12.0
    taxe_fonciere_mensuelle = taxe_fonciere / 12.0

    # Cas 1 : tout inclus (fiscalitÃ© + taxe fonciÃ¨re)
    cout_hors_credit_avec_fiscalite = (
        charges_mensuelles + depenses_annexes_mensuelles + cout_fiscal_mensuel
    )

    # Cas 2 : hors fiscalitÃ© et sans taxe fonciÃ¨re
    cout_hors_credit_sans_fiscalite = (
        charges_mensuelles + depenses_annexes_mensuelles - taxe_fonciere_mensuelle
    )

    # Cas 3 : hors fiscalitÃ©, sans taxe fonciÃ¨re, et 50 % des charges rÃ©cupÃ©rables
    cout_hors_credit_recup_50 = (
        (charges_mensuelles * 0.5) + (depenses_annexes_mensuelles - taxe_fonciere_mensuelle)
    )

    # --- Fonction gÃ©nÃ©rique de calcul de mensualitÃ© totale ---
    def mensualite_totale(prix, mode="avec_fiscalite"):
        capital_emprunte_test = prix - apport
        taux_mensuel_test = taux_annuel / 12
        nb_mois_test = duree_annees * 12
        mensualite_credit_test = capital_emprunte_test * (
            taux_mensuel_test / (1 - (1 + taux_mensuel_test) ** -nb_mois_test)
        )
        mensualite_assurance_test = (capital_emprunte_test * assurance_credit) / 12

        if mode == "avec_fiscalite":
            cout_hors_credit = cout_hors_credit_avec_fiscalite
        elif mode == "sans_fiscalite":
            cout_hors_credit = cout_hors_credit_sans_fiscalite
        elif mode == "recup_50":
            cout_hors_credit = cout_hors_credit_recup_50
        else:
            cout_hors_credit = 0

        return mensualite_credit_test + mensualite_assurance_test + cout_hors_credit

    def recherche_equilibre(mode="avec_fiscalite"):
        """Recherche dichotomique du prix dâ€™Ã©quilibre."""
        prix_min, prix_max = 0, prix_bien * 2
        for _ in range(100):
            test_prix = (prix_min + prix_max) / 2
            if mensualite_totale(test_prix, mode) > loyer_mensuel_net:
                prix_max = test_prix
            else:
                prix_min = test_prix
        return test_prix

    # --- Calcul des trois Ã©quilibres ---
    prix_equilibre_avec_fiscalite = recherche_equilibre("avec_fiscalite")
    prix_equilibre_sans_fiscalite = recherche_equilibre("sans_fiscalite")
    prix_equilibre_recup_50 = recherche_equilibre("recup_50")

    # --- Charges restantes Ã  la charge du propriÃ©taire ---
    charge_restante_sans_fisc = (charges_bien + depenses_annexes - taxe_fonciere)  # sans rÃ©cup
    charge_restante_recup_50 = ((charges_bien * 0.5) + depenses_annexes - taxe_fonciere)

    # --- AFFICHAGE DES RÃ‰SULTATS ---
    print("\n=================== ğŸ’° RÃ‰SULTATS DES Ã‰QUILIBRES ===================")

    print("\nğŸ’¡ Ã‰QUILIBRE COMPLET (avec impÃ´ts et taxes) :")
    print(f"   â¤ Prix dâ€™achat maximum : {prix_equilibre_avec_fiscalite:,.0f} â‚¬")
    print(f"     (le loyer couvre crÃ©dit + charges + impÃ´ts + taxe fonciÃ¨re)")

    print("\nğŸ’¡ Ã‰QUILIBRE HORS FISCALITÃ‰ (sans impÃ´ts ni taxe fonciÃ¨re) :")
    print(f"   â¤ Prix dâ€™achat maximum : {prix_equilibre_sans_fiscalite:,.0f} â‚¬")
    print(f"     â†’ Montant restant Ã  la charge du propriÃ©taire : {charge_restante_sans_fisc:,.0f} â‚¬/an ({charge_restante_sans_fisc/12.0:,.2f} â‚¬/mois)")
    print(f"     (le loyer couvre crÃ©dit + charges uniquement, sans impÃ´ts ni taxe fonciÃ¨re)")

    print("\nğŸ’¡ Ã‰QUILIBRE HORS FISCALITÃ‰ + 50 % CHARGES RÃ‰CUPÃ‰RABLES :")
    print(f"   â¤ Prix dâ€™achat maximum : {prix_equilibre_recup_50:,.0f} â‚¬")
    print(f"     â†’ Montant restant Ã  la charge du propriÃ©taire : {charge_restante_recup_50:,.0f} â‚¬/an ({charge_restante_recup_50/12.0:,.2f} â‚¬/mois)")
    print(f"     (le loyer couvre crÃ©dit + 50 % des charges, sans impÃ´ts ni taxe fonciÃ¨re)")

else:
    print("â„¹ï¸ Ce calcul nâ€™est possible que si le bien est financÃ© par un crÃ©dit immobilier.")
```

```python
# ================================================================
# ğŸ‡«ğŸ‡· SIMULATEUR SCI IMMOBILIÃˆRE â€” VERSION 2025 (IR / IS)
# ================================================================
# ğŸ§¾ SOURCES OFFICIELLES :
# - ImpÃ´t sur le revenu 2025 : https://www.service-public.fr/particuliers/vosdroits/F1419
# - DÃ©ficit foncier : https://www.economie.gouv.fr/particuliers/tout-savoir-deficit-foncier
# - BOFiP RFPI-BASE-30-20 : https://bofip.impots.gouv.fr/bofip/4142-PGP.html
# - PrÃ©lÃ¨vements sociaux 17,2 % : https://www.impots.gouv.fr/particulier/questions/je-donne-un-bien-en-location-dois-je-payer-des-prelevements-sociaux
# ================================================================
# âš ï¸ AVERTISSEMENTS :
# - Les calculs sont indicatifs, Ã  but pÃ©dagogique.
# - Les intÃ©rÃªts dâ€™emprunt sont estimÃ©s sur la durÃ©e moyenne du prÃªt.
# - Les rÃ©sultats ne constituent pas un avis fiscal.
# ================================================================

import math
from datetime import datetime

print("=== ğŸ‡«ğŸ‡· SIMULATEUR SCI â€” VERSION 2025 IR / IS ===\n")

# ---------------------------
# Ã‰TAPE 1 â€” FINANCEMENT
# ---------------------------
print("ğŸ¦ FINANCEMENT DU BIEN")
print("â¡ï¸ Indique si ton bien est financÃ© avec ou sans crÃ©dit immobilier.")

avec_credit = input("ğŸ‘‰ As-tu un crÃ©dit immobilier pour ce bien ? (oui/non) : ").strip().lower()

mensualite_credit = 0.0
mensualite_assurance_credit = 0.0
interets_emprunt = 0.0

if avec_credit == "oui":
    print("\nğŸ’¬ Donne les dÃ©tails de ton financement :")
    prix_bien = float(input("ğŸ‘‰ Prix dâ€™achat du bien (â‚¬) : "))
    apport = float(input("ğŸ‘‰ Montant de ton apport personnel (â‚¬) : "))
    taux_annuel = float(input("ğŸ‘‰ Taux dâ€™intÃ©rÃªt annuel du prÃªt (%) : ")) / 100
    duree_annees = int(input("ğŸ‘‰ DurÃ©e totale du prÃªt (en annÃ©es) : "))
    assurance_credit = float(input("ğŸ‘‰ Taux dâ€™assurance du prÃªt (%/an du capital empruntÃ©) : ")) / 100

    capital_emprunte = prix_bien - apport
    taux_mensuel = taux_annuel / 12
    nb_mois = duree_annees * 12

    mensualite_credit = capital_emprunte * (taux_mensuel / (1 - (1 + taux_mensuel) ** -nb_mois))
    mensualite_assurance_credit = (capital_emprunte * assurance_credit) / 12
    interets_emprunt = capital_emprunte * taux_annuel / 2  # approximation moyenne

    print(f"\nğŸ’¶ Montant empruntÃ© : {capital_emprunte:,.0f} â‚¬")
    print(f"ğŸ’³ MensualitÃ© de crÃ©dit : {mensualite_credit:,.2f} â‚¬ / mois ({mensualite_credit * 12:,.2f} â‚¬ / an)")
    print(f"ğŸ›¡ï¸ Assurance emprunteur : {mensualite_assurance_credit:,.2f} â‚¬ / mois ({mensualite_assurance_credit * 12:,.2f} â‚¬ / an)\n")

else:
    print("\nğŸ’¬ Tu achÃ¨tes sans emprunt. Indique simplement le prix du bien :")
    prix_bien = float(input("ğŸ‘‰ Prix dâ€™achat du bien (â‚¬) : "))
    capital_emprunte = 0.0
    apport = prix_bien  # tout payÃ© comptant

# ---------------------------
# Ã‰TAPE 2 â€” LOYERS & CHARGES
# ---------------------------
print("\nğŸ  LOYERS & CHARGES DU BIEN")
print("â¡ï¸ Indique le loyer que la SCI percevra et les charges non rÃ©cupÃ©rables (copro, entretien, etc.).")

loyer_mensuel = float(input("ğŸ‘‰ Loyer mensuel perÃ§u (â‚¬) : "))
loyers_annuels = loyer_mensuel * 12
print(f"   â¤ Soit {loyers_annuels:,.2f} â‚¬ / an")

charges_bien = float(input("ğŸ‘‰ Charges annuelles non rÃ©cupÃ©rables (â‚¬) : "))
print(f"   â¤ Soit {charges_bien / 12:,.2f} â‚¬ / mois")

# ---------------------------
# Ã‰TAPE 3 â€” TAXE FONCIÃˆRE
# ---------------------------
print("\nğŸ™ï¸ TAXE FONCIÃˆRE")
mode_tf = input("ğŸ‘‰ Souhaites-tu entrer un montant exact (1) ou faire une estimation (2) ? : ").strip()

if mode_tf == "1":
    taxe_fonciere = float(input("ğŸ‘‰ Montant annuel exact de la taxe fonciÃ¨re (â‚¬) : "))
else:
    print("\nğŸ’¬ Estimation simplifiÃ©e :")
    surface = float(input("ğŸ‘‰ Surface habitable du bien (mÂ²) : "))
    type_bien = input("ğŸ‘‰ Type de bien ('appartement' ou 'maison') : ").strip().lower()
    taux_communal = float(input("ğŸ‘‰ Taux global de taxe fonciÃ¨re (%) : ")) / 100
    tarif_m2 = 150 if type_bien == "appartement" else 130
    taxe_fonciere = surface * tarif_m2 * 0.5 * taux_communal

print(f"ğŸ’° Taxe fonciÃ¨re : {taxe_fonciere:,.2f} â‚¬ / an ({taxe_fonciere / 12:,.2f} â‚¬ / mois)\n")

# ---------------------------
# Ã‰TAPE 4 â€” GESTION & ASSURANCE
# ---------------------------
print("\nğŸ¢ GESTION & ASSURANCE")
gestion_agence = input("ğŸ‘‰ Gestion locative confiÃ©e Ã  une agence ? (oui/non) : ").strip().lower()
frais_agence_annuels = 0.0

if gestion_agence == "oui":
    frais_agence_pct = float(input("ğŸ‘‰ Taux de frais de gestion de lâ€™agence (%) sur les loyers : ")) / 100
    frais_agence_annuels = loyers_annuels * frais_agence_pct
    pno_incluse = input("ğŸ‘‰ Lâ€™agence inclut-elle une assurance PNO ? (oui/non) : ").strip().lower()
    if pno_incluse == "oui":
        assurance_bien = 0.0
    else:
        assurance_bien = float(input("ğŸ‘‰ Montant annuel de la PNO (â‚¬) : "))
else:
    assurance_bien = float(input("ğŸ‘‰ Montant annuel de la PNO (â‚¬) : "))

autres_depenses = float(input("ğŸ‘‰ Autres dÃ©penses diverses (entretien, comptable, etc.) (â‚¬) : "))
depenses_annexes = taxe_fonciere + assurance_bien + autres_depenses + frais_agence_annuels
print(f"ğŸ’¸ DÃ©penses annexes totales : {depenses_annexes:,.2f} â‚¬ / an ({depenses_annexes / 12:,.2f} â‚¬ / mois)\n")

# ---------------------------
# Ã‰TAPE 5 â€” RÃ‰GIME FISCAL SCI
# ---------------------------
print("\nğŸ›ï¸ RÃ‰GIME FISCAL DE LA SCI")
regime_fiscal = input("ğŸ‘‰ RÃ©gime fiscal ('ir' ou 'is') : ").strip().lower()

# ---------------------------
# Ã‰TAPE 6 â€” CALCUL FISCAL
# ---------------------------
print("\nğŸ“Š CALCUL FISCAL SELON LE RÃ‰GIME")

if regime_fiscal == "is":
    print("\nğŸ’¬ SCI Ã  lâ€™IS : indique les paramÃ¨tres comptables du bien.")
    taux_is = float(input("ğŸ‘‰ Taux dâ€™IS (%) (ex : 15 ou 25) : ")) / 100
    part_terrain = float(input("ğŸ‘‰ Pourcentage du prix correspondant au terrain (15 Ã  20 %) : ")) / 100
    duree_amortissement = int(input("ğŸ‘‰ DurÃ©e dâ€™amortissement du bÃ¢ti (annÃ©es, ex : 30 Ã  50) : "))

    base_amortissable = prix_bien * (1 - part_terrain)
    amortissement_annuel = base_amortissable / duree_amortissement

    benefice_imposable = loyers_annuels - (
        charges_bien + depenses_annexes + interets_emprunt + amortissement_annuel
    )

    cout_fiscal_total = max(0, benefice_imposable * taux_is)
    revenu_locatif_imposable = benefice_imposable
    prelevements_sociaux = 0.0
    surcout_ir = 0.0

else:
    print("\nğŸ’¬ SCI Ã  lâ€™IR : indique ton revenu personnel et les travaux dÃ©ductibles.")
    revenu_net = float(input("ğŸ‘‰ Revenu NET imposable hors SCI (â‚¬) : "))
    parts = float(input("ğŸ‘‰ Nombre de parts fiscales du foyer : "))
    travaux_deductibles = float(input("ğŸ‘‰ Montant des travaux de rÃ©paration/amÃ©lioration (â‚¬) : "))
    travaux_energetiques = input("ğŸ‘‰ Travaux Ã©nergÃ©tiques (oui/non) : ").strip().lower()

    plafond_deficit = 21400 if travaux_energetiques == "oui" else 10700

    autres_charges_deductibles = charges_bien + depenses_annexes + travaux_deductibles
    revenu_locatif_imposable = loyers_annuels - (interets_emprunt + autres_charges_deductibles)

    if revenu_locatif_imposable < 0:
        deficit_total = abs(revenu_locatif_imposable)
        imputable_global = min(deficit_total, plafond_deficit)
        revenu_locatif_imposable = 0
        deficit_foncier = imputable_global
        print(f"âš ï¸ DÃ©ficit foncier total : {deficit_total:,.0f} â‚¬")
        print(f"â†’ Imputation sur revenu global : {imputable_global:,.0f} â‚¬")
    else:
        deficit_foncier = 0

    revenu_total = revenu_net + max(0, revenu_locatif_imposable) - deficit_foncier

    def impot_progressif(revenu, parts):
        tranches = [
            (0, 11294, 0.0),
            (11294, 28797, 0.11),
            (28797, 82341, 0.30),
            (82341, 177106, 0.41),
            (177106, math.inf, 0.45),
        ]
        rp = revenu / parts
        imp = 0
        for bas, haut, taux in tranches:
            if rp > bas:
                imp += (min(rp, haut) - bas) * taux
            else:
                break
        return imp * parts

    impot_avant = impot_progressif(revenu_net, parts)
    impot_apres = impot_progressif(revenu_total, parts)
    surcout_ir = max(0, impot_apres - impot_avant)
    prelevements_sociaux = max(0, revenu_locatif_imposable * 0.172)
    cout_fiscal_total = surcout_ir + prelevements_sociaux

# ---------------------------
# Ã‰TAPE 7 â€” SYNTHÃˆSE
# ---------------------------
cout_mensuel_total = (
    mensualite_credit
    + mensualite_assurance_credit
    + (charges_bien / 12.0)
    + (depenses_annexes / 12.0)
    + (cout_fiscal_total / 12.0)
)

print("\n=================== ğŸ’° RÃ‰SULTATS ===================")
print(f"RÃ©gime fiscal : SCI Ã  {regime_fiscal.upper()}")
print(f"Revenu locatif imposable : {revenu_locatif_imposable:,.2f} â‚¬ / an ({revenu_locatif_imposable/12:,.2f} â‚¬/mois)")
print(f"SurcoÃ»t IR : {surcout_ir:,.2f} â‚¬ / an ({surcout_ir/12:,.2f} â‚¬/mois)")
print(f"PrÃ©lÃ¨vements sociaux : {prelevements_sociaux:,.2f} â‚¬ / an ({prelevements_sociaux/12:,.2f} â‚¬/mois)")
print(f"CoÃ»t fiscal total : {cout_fiscal_total:,.2f} â‚¬ / an ({cout_fiscal_total/12.0:,.2f} â‚¬/mois)")
print(f"â¡ï¸ ğŸ’¸ CoÃ»t total : {cout_mensuel_total:,.2f} â‚¬ / mois ({cout_mensuel_total * 12:,.2f} â‚¬ / an)")

# ---------------------------
# Ã‰TAPE 8 â€” PRIX Dâ€™ACHAT MAXIMUM Ã€ Lâ€™Ã‰QUILIBRE
# ---------------------------
print("\nğŸ“ˆ CALCUL DU PRIX Dâ€™ACHAT MAXIMUM Ã€ Lâ€™Ã‰QUILIBRE")

if avec_credit == "oui" and mensualite_credit > 0:
    loyer_mensuel_net = loyer_mensuel

    # --- CoÃ»ts mensuels selon diffÃ©rents scÃ©narios ---
    cout_fiscal_mensuel = cout_fiscal_total / 12.0
    charges_mensuelles = charges_bien / 12.0
    depenses_annexes_mensuelles = depenses_annexes / 12.0
    taxe_fonciere_mensuelle = taxe_fonciere / 12.0

    # Cas 1 : tout inclus (fiscalitÃ© + taxe fonciÃ¨re)
    cout_hors_credit_avec_fiscalite = (
        charges_mensuelles + depenses_annexes_mensuelles + cout_fiscal_mensuel
    )

    # Cas 2 : hors fiscalitÃ© et sans taxe fonciÃ¨re
    cout_hors_credit_sans_fiscalite = (
        charges_mensuelles + depenses_annexes_mensuelles - taxe_fonciere_mensuelle
    )

    # Cas 3 : hors fiscalitÃ©, sans taxe fonciÃ¨re, et 50 % des charges rÃ©cupÃ©rables
    cout_hors_credit_recup_50 = (
        (charges_mensuelles * 0.5) + (depenses_annexes_mensuelles - taxe_fonciere_mensuelle)
    )

    # --- Fonction gÃ©nÃ©rique de calcul de mensualitÃ© totale ---
    def mensualite_totale(prix, mode="avec_fiscalite"):
        capital_emprunte_test = prix - apport
        taux_mensuel_test = taux_annuel / 12
        nb_mois_test = duree_annees * 12
        mensualite_credit_test = capital_emprunte_test * (
            taux_mensuel_test / (1 - (1 + taux_mensuel_test) ** -nb_mois_test)
        )
        mensualite_assurance_test = (capital_emprunte_test * assurance_credit) / 12

        if mode == "avec_fiscalite":
            cout_hors_credit = cout_hors_credit_avec_fiscalite
        elif mode == "sans_fiscalite":
            cout_hors_credit = cout_hors_credit_sans_fiscalite
        elif mode == "recup_50":
            cout_hors_credit = cout_hors_credit_recup_50
        else:
            cout_hors_credit = 0

        return mensualite_credit_test + mensualite_assurance_test + cout_hors_credit

    def recherche_equilibre(mode="avec_fiscalite"):
        """Recherche dichotomique du prix dâ€™Ã©quilibre."""
        prix_min, prix_max = 0, prix_bien * 2
        for _ in range(100):
            test_prix = (prix_min + prix_max) / 2
            if mensualite_totale(test_prix, mode) > loyer_mensuel_net:
                prix_max = test_prix
            else:
                prix_min = test_prix
        return test_prix

    # --- Calcul des trois Ã©quilibres ---
    prix_equilibre_avec_fiscalite = recherche_equilibre("avec_fiscalite")
    prix_equilibre_sans_fiscalite = recherche_equilibre("sans_fiscalite")
    prix_equilibre_recup_50 = recherche_equilibre("recup_50")

    # --- Charges restantes Ã  la charge du propriÃ©taire ---
    charge_restante_sans_fisc = (charges_bien + depenses_annexes - taxe_fonciere)  # sans rÃ©cup
    charge_restante_recup_50 = ((charges_bien * 0.5) + depenses_annexes - taxe_fonciere)

    # --- AFFICHAGE DES RÃ‰SULTATS ---
    print("\n=================== ğŸ’° RÃ‰SULTATS DES Ã‰QUILIBRES ===================")

    print("\nğŸ’¡ Ã‰QUILIBRE COMPLET (avec impÃ´ts et taxes) :")
    print(f"   â¤ Prix dâ€™achat maximum : {prix_equilibre_avec_fiscalite:,.0f} â‚¬")
    print(f"     (le loyer couvre crÃ©dit + charges + impÃ´ts + taxe fonciÃ¨re)")

    print("\nğŸ’¡ Ã‰QUILIBRE HORS FISCALITÃ‰ (sans impÃ´ts ni taxe fonciÃ¨re) :")
    print(f"   â¤ Prix dâ€™achat maximum : {prix_equilibre_sans_fiscalite:,.0f} â‚¬")
    print(f"     â†’ Montant restant Ã  la charge du propriÃ©taire : {charge_restante_sans_fisc:,.0f} â‚¬/an ({charge_restante_sans_fisc/12.0:,.2f} â‚¬/mois)")
    print(f"     (le loyer couvre crÃ©dit + charges uniquement, sans impÃ´ts ni taxe fonciÃ¨re)")

    print("\nğŸ’¡ Ã‰QUILIBRE HORS FISCALITÃ‰ + 50 % CHARGES RÃ‰CUPÃ‰RABLES :")
    print(f"   â¤ Prix dâ€™achat maximum : {prix_equilibre_recup_50:,.0f} â‚¬")
    print(f"     â†’ Montant restant Ã  la charge du propriÃ©taire : {charge_restante_recup_50:,.0f} â‚¬/an ({charge_restante_recup_50/12.0:,.2f} â‚¬/mois)")
    print(f"     (le loyer couvre crÃ©dit + 50 % des charges, sans impÃ´ts ni taxe fonciÃ¨re)")

else:
    print("â„¹ï¸ Ce calcul nâ€™est possible que si le bien est financÃ© par un crÃ©dit immobilier.")
```
